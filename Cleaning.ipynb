```python
import pandas as pd

# Define the file name.
# Replace this with the specific file you want to analyze.
file_name = "แบบสอบถามพฤติกรรมการรับชมแอนิเมชัน.csv"

# --- 1. Load the Dataset ---
# We use a try-except block to handle potential encoding issues,
# which is common for datasets with non-English characters.
try:
    df = pd.read_csv(file_name)
except Exception:
    # Try reading with a common Windows/Thai encoding
    try:
        df = pd.read_csv(file_name, encoding='windows-1252')
    except Exception:
        # Default to UTF-8
        df = pd.read_csv(file_name, encoding='utf-8')

print(f"--- Loaded DataFrame '{file_name}' ---")
print(f"Total Rows: {len(df)}")
print(f"Total Columns: {len(df.columns)}")
print("-" * 35)


# ===============================================
# 2. Check for Missing Values (NaN/Null)
# ===============================================

print("             MISSING VALUES ANALYSIS")
print("="*50)

# Calculate total count of missing values per column
missing_counts = df.isnull().sum()

# Calculate percentage of missing values per column
missing_percentages = (missing_counts / len(df)) * 100

# Combine the results into a single DataFrame for presentation
missing_summary = pd.DataFrame({
    'Missing Count': missing_counts,
    'Missing Percentage': missing_percentages.round(2).astype(str) + '%'
})

# Filter to show only columns with at least one missing value and sort
missing_summary = missing_summary[missing_summary['Missing Count'] > 0]
missing_summary = missing_summary.sort_values(by='Missing Count', ascending=False)

if missing_summary.empty:
    print("NO MISSING VALUES FOUND in any column.")
else:
    # Use to_string() for clean console formatting
    print(missing_summary.to_string())


# ===============================================
# 3. Check for Duplicate Values
# ===============================================

print("\n             DUPLICATE VALUES ANALYSIS")
print("="*50)

# Total number of duplicate rows (excluding the first occurrence)
total_duplicates = df.duplicated().sum()
print(f"Total Number of Duplicate Rows Found (excluding first instance): {total_duplicates}")

if total_duplicates > 0:
    # Get the actual duplicate rows (showing ALL instances of the duplicate set)
    duplicate_rows = df[df.duplicated(keep=False)].sort_index()

    print("\nFirst 50 Actual Duplicate Rows (showing all instances, limited for brevity):\n")
    # Use to_string() for clean console formatting
    print(duplicate_rows.head(50).to_string())
else:
    print("NO DUPLICATE ROWS FOUND.")
```
